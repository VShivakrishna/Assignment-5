```python
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.decomposition import PCA

# You can add the parameter data_home to wherever to where you want to download your data
dataset = pd.read_csv('./datasets/CC.csv')
dataset = dataset.fillna(dataset.mean(numeric_only=True))
x = dataset.iloc[:,1:17].values
y = dataset.iloc[:,17].values

pca = PCA(2)
x_pca = pca.fit_transform(x)
df2 = pd.DataFrame(data=x_pca)
finaldf = pd.concat([df2,dataset[['TENURE']]],axis=1)
finaldf.head()
```

```python
##### K-means and PCA ########
#Applying K-means algorithm
from sklearn.cluster import KMeans
nclusters = 3 
km = KMeans(n_clusters=nclusters)
km.fit(x)

# predict the cluster for each data point
y_cluster_kmeans = km.predict(x)
from sklearn import metrics
score = metrics.silhouette_score(x, y_cluster_kmeans)
print("Silhouette_score after feature scaling:",score)
```

```python
####scaling, K-means and PCA#########
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# Fit on training set only.
scaler.fit(x)

# Applying feature scaling
x_scaler = scaler.transform(x)
pca = PCA(2)
x_pca = pca.fit_transform(x_scaler)
df2 = pd.DataFrame(data=x_pca)
finaldf = pd.concat([df2,dataset[['TENURE']]],axis=1)
finaldf.head()
```

```python
from sklearn.cluster import KMeans
nclusters = 3 
km = KMeans(n_clusters=nclusters)
km.fit(x_scaler)

# predict the cluster for each data point
y_cluster_kmeans = km.predict(x_scaler)
from sklearn import metrics
score = metrics.silhouette_score(x_scaler, y_cluster_kmeans)
print("Silhouette_score after feature scaling:",score)
```
